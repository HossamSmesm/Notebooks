### here i'll select only ean columnsand apply a correlation heatmap to see tha crrelation between them clearly. 
### we can drop the feature that has a strong correlation with another couse it will do the same role in the model..
#### Ex: "area_mean" ," perimeter_mean" and "radius_mean " have 0.99 corr so we could drop two of them
#### also "concavity_mean " and "concave points_mean" so we could drop "concave points_mean"

```python
mean_df=df.iloc[:,2:12]
plt.figure(figsize=(14,7))
sns.heatmap(mean_df.corr(), annot=True)
```

```python
mean_df.drop(['perimeter_mean','area_mean','concave points_mean'], axis=1, inplace=True)
plt.figure(figsize=(14,7))
sns.heatmap(mean_df.corr(), annot=True)
#so this looks nice :) .
```
